```python
# 1. Simple Logging Decorator
# Write a decorator log_call that prints Function called each time the decorated
# function is executed.
def log_call(func):
    def wrapper(*args, **kwargs):
        print("Function called")
        return func(*args, **kwargs)
    return wrapper

@log_call    
def greet():
    print("Hello!")

@log_call
def add(a, b):
    return a + b

greet()
print(add(2,3))
```

```python
# 2Ô∏è‚É£ Uppercase Return Decorator
# Create a decorator uppercase that takes a function returning a string and returns
# the same string in uppercase.
def upperCase(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result.upper()
    return wrapper

@upperCase
def hello(user):
    return f"Hello {user}"

hello("Wifi")
```

```python
# 3Ô∏è‚É£ Timing Decorator (OUT OF SYLLABUS)
# Write a decorator timer that measures how long a function takes to execute and
# prints the duration.
from time import(time, sleep)
def timeMajor(func):
    def wrapper():
        start = time()
        (func())
        end = time()
        print(f"Total time taken to execute the function:{round((end - start),2)} second")
        # return result
    return wrapper

@timeMajor
def hello():
    sleep(1)
    return print("Hello Done")

@timeMajor
def say_hi():
    sleep(2)
    print("Hi done")

hello()
say_hi()
```

```python
# 4Ô∏è‚É£ Result Doubling Decorator
# Write a decorator double_result that doubles the numeric result of the decorated
# function.
def double_result(func):
    def wrapper(*args, **kwrgs):
        result = func(*args, **kwrgs)
        return result * 2
    return wrapper

@double_result
def add(*args):
    return sum(args)

print("Output:",add(2,3))
```

```python
# 5Ô∏è‚É£ No Return Decorator
# Create a decorator announce that prints "Running..." before the function runs but
# doesn‚Äôt modify the return value.
def announce(func):
    def wrapper():
        print("Running...")
        return func()
    return wrapper

@announce
def say_hi():
    print("Hi!")
    return "Done"
say_hi()
```

```python
# 6Ô∏è‚É£ Bold HTML Decorator
# Write a decorator make_bold that wraps the return string in <b>...</b> .

def make_bold(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        if isinstance(result, str):
            return f"<b>{result}</b></"
    return wrapper

@make_bold
def greet():
    return 12

print("Output: ",greet())
```

```python
# 7Ô∏è‚É£ Argument Logger
# Create a decorator log_args that prints all positional and keyword arguments
# passed to the function.
def log_args(func):
    def wrapper(*args, **kwargs):
        print("Positional arguments:", args)
        print("Keyword arguments:", kwargs)
        return func(*args, **kwargs)
    return wrapper

@log_args
def add(x,y):
    return x+y

print(add(4, y=5))
```

```python
def log_args(function):
    def wrapper(*args,**kwargs):
        function(*args,**kwargs)
        print(f"Arguments: {args},{kwargs}")

    return wrapper
    
@log_args 
def add(x,y):
    return x+y 

add(4,y=5)
```

```python
def decorator(func):
    def gate(*args, **kwargs):
        name = kwargs.get("name", "")
        if name == "Dinkar":
            func(*args, **kwargs)
        else:
            print('Not registered')
    return gate

def registered(name):
    print(f"{name} is registred")

# how decorator is works - working flow of decorator

gate = decorator(registered)
gate("Dinkar")
```
