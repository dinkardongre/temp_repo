```python
# **1. Merge two strings character by character (alternating)**

# **Input:** 
s1 = "abc"
s2 = "xyz"
s3 = ""
for i in range(0, len(s1)):
    s3 = s3 + s1[i] + s2[i]
    # s3 += s2[i]
s3
```

```python
# **2. Check if one string is a rotation of another**

# **Input:** 
s1 = "ashsh"
s2 = "hshas"
def rotationCheck(string1, string2):
    if len(string1) != len(string2):
        return False
    temp = string1 + string2
    if string2 in temp:
        return True
    else:
        return False


if rotationCheck(s1, s2):
    print(f"Yes, s2 is a rotation of s1:")
else :
    print("No, s2 is not a rotation of s1")
```

```python
# # **3. Replace every vowel in a string with its index position**

w = "hello world"
l = ['a','e','i','o','u']
ans = ""
for i in range(0, len(w)):
    if w[i] in l:
        ans += str(i)
    else:
        ans += w[i]
ans
```

```python
# **5. Print all palindromes from a list of words**

inp = ["madam", "racecar", "apple", "level", "banana"]
for i in inp:
    a = i[::-1]
    if a == i:
        print(a)
```

```python
# **7. Group strings by their length using a dictionary**

inp = ["sun", "sky", "python", "java", "moon"]
# a = {3: ['sun', 'sky'], 6: ['python'], 4: ['java', 'moon']}
a = {}
for i in inp:
    n = len(i)
    if n not in a:
        a[n] = [i]
    else:
        a[n].append(i)
a
```

```python
# 8. Check if all elements in a list are unique

inp = [1, 2, 3, 4, 5, 2]
inp2 = []
for i in inp:
    if i not in inp2:
        inp2.append(i)

if len(inp) != len(inp2):
    print("No, list contains duplicate values")
else:
    print("Yes, list have all unique elements")
```

```python
# **9. Find second highest number without using `sort()` or multiple `max()`**

a = [10, 25, 30, 45, 22, 44, 46]
max = 0
smax = 0
for i in a:
    if max < i:
        smax = max
        max = i
    elif smax < i:
        smax = i
    
smax
```

```python
# **13. Merge two dictionaries, summing values for common keys**

# **Input:**

dict1 = {'a': 10, 'b': 20}

dict2 = {'b': 5, 'c': 15}

for i,v in dict2.items():
    if i in dict1:
        dict1[i] += v
    else:
        dict1[i] = v

dict1
```

```python
# **14. From a list, group elements by their type into a dictionary**

List = [1, "hello", 3.5, "world", 42, 0.0]
wr = {}

for i in List:
    j = type(i)
    if j not in wr:
        wr[j] = [i]
    else :
        wr[j].append(i)
wr
```

```python
# **15. Reverse the keys and values in a dictionary**

dic = {"apple": "red", "banana": "yellow"}
i = list(dic.items())
for k,v in i:
    dic[v] = k
    dic.pop(k)
dic
```

```python
# **16. Sort a dictionary by values in descending order**

hf = {"Math": "75", "English": 85, "Science": 80}
hf = list(hf.items())
def srt(tup):
    print(tup)
    return int(tup[1])

sorted(hf, key = srt, reverse=True)
```
