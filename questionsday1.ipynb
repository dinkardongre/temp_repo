```python
def palindromCheck(srt):
    srt = srt.lower()
    i = 0
    j = len(srt)-1

    while(i != j):
        if srt[i] != srt[j]:
            return False
        i += 1
        j -= 1
    return True

palindromCheck("Madam")
```

```python
# 4ï¸âƒ£ Second Largest Element
arr = [10, 20, 4, 45, 99]

def secLar(arr):
    largest = 0
    slargest = 0
    for i in arr:
        if i > largest:
            slargest = largest
            largest = i
        elif i > slargest and i < largest:
            slargest = i
    return slargest

print("Second largest: ", secLar(arr))
```

```python
# Reverse a number
num = 4543
rev = 0
while(num != 0):
    ld = num % 10
    rev = rev*10 + ld
    num //= 10

rev
```

```python
# Check if a number is a palindrome
num = 454
rev = 0
tempnum = num
while(tempnum != 0):
    ld = tempnum % 10
    rev = rev*10 + ld
    tempnum //= 10

num == rev
```

```python
# factorial
n = 5
fact = 1
for i in range(1, n+1):
    fact *= i
print(fact)

def fact(n):
    if n == 0 or n == 1:
        return 1
    return n * fact(n-1)

print(fact(n))
```

```python
# Check if a number is prime
n = 5

is_prime = True

if n <= 1:
    is_prime = False
else:
    for i in range(2, n):
        if n % i == 0:
            is_prime = False
            break

if is_prime is True:
    print ("Prime Number")
else:
    print("Not a prime number")
```

```python
def fabb(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    return fabb(n-1) + fabb(n-2)

n = 6
sum = 0
for i in range(n):
    print(fabb(i), end=" ")
    sum += fabb(i)
print("Sum of fabb till n", sum)
```
