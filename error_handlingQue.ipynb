```python
# 7.	Write code to demonstrate the use of finally for closing a file, even if an error occurs.
try:
    file = open("example.txt", "r")   # Try to open file
    content = file.read()
    print("File content:\n", content)

    # Force an error for demo
    num = 10 / 0   

except ZeroDivisionError:
    print("⚠️ Division by zero occurred!")

except FileNotFoundError:
    print("⚠️ The file does not exist.")

finally:
    # This block always runs
    try:
        file.close()
        print("✅ File is closed (even if an error occurred).")
    except NameError:
        print("⚠️ File was never opened, so nothing to close.")
```

```python
# 19.	Create a decorator that catches exceptions and prints “An error occurred” instead of crashing.
def safe_run(func):
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except Exception:
            print("An error occurred")
    return wrapper

@safe_run
def divide(a, b):
    return a / b

print(divide(10, 2))
print(divide(10, 0))
```

```python
	# 20.	Implement a program where you intentionally trigger and handle a custom-defined exception class (e.g., InvalidAgeError).
class InvalidAgeError(Exception):
    def __init__(self, age, message="Age must be between 18 and 60"):
        self.age = age
        self.message = message
        super().__init__(self.message)
try:
    age = int(input("Enter your age: "))
    if age < 18 or age > 60:
        raise InvalidAgeError(age)
    else:
        print("Valid age! You can proceed.")
        
except InvalidAgeError as e:
    print(f"InvalidAgeError: {e.message}. You entered {e.age}.")
    
except ValueError:
    print("Please enter a valid number for age.")
```
