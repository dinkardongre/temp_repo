```python
# Q1. Square Each Number
# Given a list of numbers, use map() to return a list of their squares.
inp = [1,2,3,4]
def hello(num):
    return num**2
list(map(hello, inp))
```

```python
# Q3. Length of EachWord
# Given a list of words, return the length of each using map() .
ip = ["apple", "orange", "banana"]
def hen(name):
    return len(name)
list(map(hen, ip))
```

```python
# Q4. Celsius to Fahrenheit
# Convert a list of Celsius temperatures to Fahrenheit using map() .
tempInCel = [0, 20, 37]
def converter(c):
    return f"{c} Celsius in Fahrenheit {c * 9/5 + 32}"
list(map(converter, tempInCel))
```

```python
# Q5. Append a Suffix
# Given a list of names, use map() to add "@gmail.com" to each.
names = ["nikita", "khushi"]
def inp(name):
    return name+"@gmail.com"
list(map(inp, names))
```

```python
# Q6. Filter Even Numbers
# Use filter() to extract even numbers from a list.
inp =  [1, 2, 3, 4, 5, 6]
def evenNum(num):
    return num % 2 == 0
list(filter(evenNum, inp))
```

```python
# Q7. FilterWords Longer Than 4 Letters
# Given a list of words, keep only those longer than 4 letters
words = ["Hello", "woekd", "gaps", "abv"]
def ml(name):
    if len(name) > 4:
        return name
tuple(filter(ml, words))
```

```python
# Q8. Filter Positive Numbers
# From a list of integers, keep only the positive ones using filter() .
Input = [-3, 0, 5, -1, 7]
def checker(num):
    if num > 0:
        return num

list(filter(checker, Input)) 
```

```python
# Q9. Filter Palindromes
# Use filter() to find all palindromes in a list of strings.
lst = ["madam", "nayan", "eye", "ashish"]
def palindrome(name):
    i = 0
    j = len(name)-1
    while(i < j):
        if name[i] != name[j]:
            return False
        i+=1
        j-=1
    return True
list(filter(palindrome, lst))
```

```python
# Q10. Filter Multiples of 3
# From a list, filter numbers divisible by 3.
lst = [12,43,15,18,9,67]
def hello(num):
    return num % 3 == 0
list(filter(hello, lst))
```

```python
# 30. Rotate the array to the right by k steps.
#     Input: arr = [1, 2, 3, 4, 5, 6, 7], k = 3
#     Output: [5, 6, 7, 1, 2, 3, 4]``

# Note: Rotation means last k elements come to the front

arr = [1,2,3,4,5,6,7]
k = 3
arr1 = []
for i in range(len(arr)-1, len(arr)-k-1, -1):
    arr1.append(arr[i])
    
for i in range(0, len(arr)-k):
    arr1.append(arr[i])

arr1
```
