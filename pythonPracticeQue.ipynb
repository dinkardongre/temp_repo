```python
### Q7. Print Fibonacci Sequence Till N

# **Problem:** Print Fibonacci numbers less than or equal to N using a while loop.
n = 5
i = 0
j = 1
a = 0
# print(i, end=" ")
# print(j, end = " ")
isPrime = lambda num: len ([ j for j in range(2, num) if num % j == 0]) == 0
while(a < n):
    k = i + j
    if isPrime(k) and k > 1:
        print(k, end = " ")
        a += 1
    i = j
    j = k
```

```python
### Q8. Check if a Number is Palindrome

# **Problem:** Check whether a number is a palindrome using a while loop.
N = 121
N = str(N)
i = 0
j = len(N)-1
is_pali = True
while(i < j):
    if N[i] != N[j]:
        is_pali = False
        break
    i += 1
    j -= 1

# N = int(N)
if is_pali:
    print(N,"Palindrome")
else:
    print("Not Palindrome")
```

```python
### Q2. Password Strength Checker

# **Problem:** Write a function `check_password(password)` that checks:

# - Minimum 8 characters
# - Contains at least one uppercase, one lowercase, one digit, one special character
def check_password(password):
    lenCount = 0
    upCount = 0
    loCount = 0
    dCount = 0
    spcCount = 0
    if len(password)>=8:
        lenCount +=1
    for i in password:
        if i.isdigit():
            dCount += 1
        if i.islower():
            loCount += 1
        if i.isupper():
            upCount += 1
        if not i.isalnum():
            spcCount += 1
    if lenCount > 0 and upCount > 0 and loCount > 0 and dCount > 0 and spcCount > 0:
        return print("Password is very strong")
    # if lenCount > 0 or upCount > 0 or loCount > 0 or dCount > 0 or spcCount > 0:
    #     return "Password is not so strong"

    return print("Password is weak")

check_password("Abc123!")
check_password("StrongPass1$")
```

```python
def check_password(password):
    # Conditions
    length_ok = len(password) >= 8
    has_upper = any(ch.isupper() for ch in password)
    has_lower = any(ch.islower() for ch in password)
    has_digit = any(ch.isdigit() for ch in password)
    has_special = any(not ch.isalnum() for ch in password)  # non-alphanumeric

    if all([length_ok, has_upper, has_lower, has_digit, has_special]):
        return "Password is very strong"
    else:
        return "Password is weak"

# Testing
print(check_password("Abc123!"))       # Weak (too short, 7 chars)
print(check_password("StrongPass1$"))  # Very strong
```

```python
### Q4. Quiz Game Scorer

# **Problem:** Write a function `quiz_score(answers, key)` that takes a list of answers and the correct key, and returns the score. Each correct answer = 1 point.
answers = ["A", "B", "C", "D"]
keys = ["A", "C", "C", "D"]

def quiz_score(answers, key):
    score = 0
    for i in range(0, len(answers)):
        if answers[i] == keys[i]:
            score += 1
    return score

quiz_score(answers, keys)

```

```python
### Q5. Shopping Cart Total

# **Problem:** Write a function `calculate_total(cart)` where `cart` is a list of dictionaries with `"item"`, `"price"`, `"quantity"`. Return total bill.
cart = [
    {"item": "Apple", "price": 50, "quantity": 2},
    {"item": "Banana", "price": 10, "quantity": 5}
]
def calculate_total(cart):
    total_bill = 0
    for i in cart:
        total_bill += i["price"]*i["quantity"]
    return total_bill

calculate_total(cart)
```

```python
### Q6. Discount Finder

# **Problem:** Write a function `apply_discount(price, customer_type)` that applies:

# - `"Regular"` ‚Üí 5%
# - `"Premium"` ‚Üí 10%
# - `"VIP"` ‚Üí 20%
def apply_discount(price, customer_type = "Nothing"):
    if customer_type == "Premium":
        return price-(price*.1)
    elif customer_type == "Regular":
        return price-(price*.05)
    elif customer_type == "VIP":
        return price-(price*.2)
    else:
        return "No discount",price
    
apply_discount(1000, "Premium")
# apply_discount(500, "Regular")
```

```python
### Q10. Simple Quiz ‚ÄúGuess the Number‚Äù

# **Problem:** Write a function `guess_number(secret, guesses)` that takes a secret number and a list of guesses. It prints `"Too high"`, `"Too low"`, or `"Correct"` for each guess.
def guess_number(secret, guesses):
    for i in guesses:
        if i < secret:
            print("Too low")
        elif i > secret:
            print("Too high")
        else:
            print("Correct")

guess_number(7, [5,8,7,9,3,7])
```
