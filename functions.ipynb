```python
def Sq(x):
    if x % 2 == 0:
        return x**2
    return x
li = (2,3,4,5,6,7,8,9)
m = tuple(map(Sq, li))
m
```

```python
def Up(x):
    return x[0].upper()+x[1:(len(x))]

li = "hello dinkar give me some monye"
li = li.split(" ")
li = list(map(Up, li))
li
```

```python
# -- filter -- return True or False then execute the data according to true or false
def odd(x):
    return x%2 != 0
li = {2,3,4,5,6,7,8,9}
m = list(filter(odd, li))
m
```

```python
friends = [
        {"name":"A", "age":9},
        {"name":"B", "age":2},
        {"name":"C", "age":7},
        {"name":"D", "age":1},
        {"name":"E", "age":6},
]
def sF(oneData):
    return oneData["age"]
sorted(friends, key = sF, reverse=-1)
```

```python
data = [("a", 2), {"b": 3}, {"c": 4}]

def to_dict(item):
    if isinstance(item, tuple):
        return {item[0]: item}
    elif isinstance(item, dict):
        return item
    return {}

li = list(map(to_dict, data))
li
```

```python
data = [("a", 2), {"b": 3}, {"c": 4}]

def rt(t):
    k = t[0]
    v = t[1]
    return {k,v}

list(map(rt, data))
```
