```python
# ✅ Keyword & Positional Practice
# 1. Write a function that takes item , price , and quantity , and prints the total cost.
def fruitsMarket(item, price, quantity):
    return f"The {item} have a price of {price} Ruppes per kg you want {quantity} so it will cost you {price * quantity}"
fruitsMarket("Apple", 120, 2)
print(fruitsMarket(item="Sugar", quantity=10, price=40))  # it is called keyword argument
```

```python
# 2. Make a function that takes principal , rate , and time and calculates simple
# interest.
def SI(P,R,T = 1):
    return (P * R * T)/100
SI(1000, 5, 9)
```

```python
# 1. Create a function that prints a sentence like:
# Function Questions 1
# "Order: 2 Pizza(s), 1 Coke(s)"
# pizza=1 , coke=0 should be default
# Allow the user to pass different values
def order(pizza = 1, coke = 0):
    return f"Order: {pizza} Pizza(s), {coke} Coke(s)"
order(2, 1)
```

```python
# 1. Write a function to calculate the final price after discount. It should take 
# price , discount_percent=10 , and a tax_percent=5 .
# Call it with different combinations of arguments.
def dis(price, discount_percent = 10, tax_percent = 5):
    return f"Final price after discount: {price - (price*(discount_percent + tax_percent))/100}"
dis(10)
```

```python
# 2. Create a function that accepts 3 numbers and returns the largest one. Call it
# in different ways:
# Only positional
# Mixed positional and keyword
# All keyword arguments
def largest(num1, num2, num3):
    if num1 > num2 and num1 > num2:
        return f'The largest {num1}'
    elif num2 > num3:
        return f'The largest {num2}'
    else:
        return f'The largest {num3}'
largest(10,32,43)
```
