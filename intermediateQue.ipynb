```python
### **1. Count vowels and consonants in a string**

Input = "Education"
v = 0
c = 0
for i in Input:
    i = i.lower()
    if i == 'a' or i == 'e' or i == 'i' or i == 'o' or i == 'u':
        v += 1
    else :
        c+=1

print(f"Vowels are {v} in string")
print(f"Consonants are {c} in string")
```

```python
# üßÆ 6. Count character frequency in a string
s = "apple"
freq = {}

for char in s:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1    

print(freq)
```

```python
### 21. Count how many vowels are in a string using `while` loop**

# Use a loop to go through the string character by character and count vowels.
strr = "education"
countV = 0
i = 0
vowels = ['a', 'e','i','o','u']
while (i < len(strr)):
    if strr[i] in vowels :
        countV += 1
    i += 1

countV
```

```python
### 22. Multiply digits of a number using `while` loop**

# Use a loop to extract each digit and multiply them.
num = 234
ml = 1
while(num > 0):
    ld = num % 10
    ml = ml * ld
    num //= 10

print(f"Product of digits is: {ml}")
```

```python
### 23. Count how many times digit 5 appears in a number**

# Use a loop to extract digits and count only if the digit is 5.
num = 55345
count = 0
while(num > 0):
    ld = num % 10
    if ld == 5:
        count += 1
    
    num //= 10

print(f"Digit 5 appears {count} times")
```

```python
### 28 Merge two dictionaries. If keys overlap, add their values**

# **Input:**

d1 = {'a': 10, 'b': 20}
d2 = {'b': 5, 'c': 15}

for key, value in d2.items():
    if key in d1:
        d1[key] += value
    else:
        d1[key] = value

print(d1)
```
