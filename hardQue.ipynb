```python
words = ["radar", "apple", "level"]
print(list(filter(lambda w: w == w[::-1], words)))
```

```python
data = ["python", "java", "c++", "ruby"]
res = list(map(lambda x: x[::-1], filter(lambda x: len(x) > 3, data)))
print(res)
```

```python
nums = [1, 2, 3, 4, 5]
result = list(filter(lambda x: x % 2 == 0, map(lambda x: x**2, nums)))
print(result)
```

```python
def mystery(x, lst=[]):
    lst.append(x)
    return lst

print(mystery(1))
print(mystery(2))
print(mystery(3, []))
print(mystery(4))
```
