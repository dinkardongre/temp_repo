```python
### 38. Print numbers from 1 to 5 using a while loop**

# Use a `while` loop to print numbers starting from 1 up to 5.
Start = 1
End = 5
while(Start <= End):
    print(Start)
    Start +=1
```

```python
### 39. Print the reverse of a number using a while loop**

# Use a loop to reverse the digits of the number.
num = 4321
# shortcut
# rev = int(str(num)[::-1]) # 1234
num1 = 0
while(num != 0):
    lastDigit = num % 10
    num1 = (num1* 10) + lastDigit
    num //=10

num1
```

```python
### 40. Calculate the sum of digits of a number using while loop**

# Use a loop to extract digits and add them.
num = 1234
sum = 0
while(num != 0):
    sum += num%10
    num//=10

print("Sum of digits:", sum)
```

```python
### 41. Print all even numbers from 10 to 2 in reverse using while loop**

# Use a loop to decrement and check if the number is even.
start = 10
end = 2
while(end <= start):
    if start % 2 == 0:
        print(start)
    start -= 1
```

```python
### 42. Count how many digits are there in a number using while loop**

# Use a counter inside a loop to count digits.
num = 12345
td = 0
while(num != 0):
    td += 1
    num//= 10

print("Total digits =", td)
```
