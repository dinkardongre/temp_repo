```python
# 3. Average of list number and their closest number to avg
num = [1, 2, 5, 7, 3, 16, 20, 43]
avg = int(sum(num)/len(num))
closest = num[0]
min_gape = abs(avg - num[0])
for i in num :
    if abs(avg - i) < min_gape :
        min_gape = abs(avg - i)
        closest = i
print("Average:", avg)
print("Closest number to average:", closest)
```

```python
# 3. A bus has 10 seats.
# Ask the user for number of seats they want to book.
# If seats available, book them. If not, show “Only X seats left”.
# Bus seating data
busAv = [
    {"seat": 1, "status": "Available"},
    {"seat": 2, "status": "NotAvailable"},
    {"seat": 3, "status": "NotAvailable"},
    {"seat": 4, "status": "Available"},
    {"seat": 5, "status": "NotAvailable"},
    {"seat": 6, "status": "Available"},
    {"seat": 7, "status": "Available"},
    {"seat": 8, "status": "Available"},
    {"seat": 9, "status": "Available"},
    {"seat": 10, "status": "NotAvailable"}
]
available_seats = [seat for seat in busAv if seat["status"] == "Available"]
num_available = len(available_seats)

seats_to_book = int(input("How many seats do you want to book? "))

if seats_to_book <= num_available:
    print("Booking ",seats_to_book," seat(s)... Booked successfully!")
else:
    print("Only", num_available ,"seat(s) left.")
```

```python
# 4. A user inputs number of days the book is late.
# Late fee: ₹2 per day for first 5 days, ₹5 for next 5 days, ₹10 beyond that.
# Calculate and print total fine.
daysLate = int(input())
totalFine = 0
for i in range(1, daysLate+1):
    if i == 1 or i == 2 or i == 3 or i == 4 or i == 5 :
        totalFine += 2
    elif i == 6 or i == 7 or i == 8 or i == 9 or i == 10 :
        totalFine += 5
    else :
        totalFine += 10
totalFine
```
